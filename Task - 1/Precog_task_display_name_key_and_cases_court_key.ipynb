```python
df2 = pd.read_csv('/content/drive/Shareddrives/Precog/Precog/Key/cases_court_key.csv')
df2.head()
```

```python
df_state_wise = df2.groupby(['state_name','year'])['state_code'].value_counts().to_frame()
df_state_wise = df_state_wise.rename(columns={'state_code':'count'})
# df_state_wise_final = pd.DataFrame(columns={"state":df_state_wise.index,"count":df_state_wise.values})
df_state_wise = df_state_wise.reset_index(['state_name','year'])
df_state_wise.head()
```

```python
fig2 = px.line(df_state_wise, x='year', y='count',title='Number of Courts : 2010-2018', color='state_name', symbol="state_name")
fig2.show()
```
