**CASES CLOSED PER STATE**

```python
df_status_count = df.groupby(['state_name'])['status'].value_counts().to_frame()
df_status_count = df_status_count.rename(columns={'status':'count'})
df_status_count = df_status_count.reset_index(['state_name','status'])
df_status_count.head()
```

```python
df_closed = pd.DataFrame({'state_name':df_status_count['state_name'],'count':df_status_count.loc[df_status_count['status'] == 'closed']['count']})
df_closed = df_closed.dropna(subset=['count'], how='all')
df_closed.head()
```

```python
df_closed_cases = pd.merge(map_df,df_closed,on='state_name', how='left')
df_closed_cases=df_closed_cases.drop(columns=['ID_0', 'ISO', 'NAME_0', 'ID_1', 'HASC_1', 'CCN_1',
        'CCA_1', 'TYPE_1', 'ENGTYPE_1', 'NL_NAME_1', 'VARNAME_1'])
df_closed_cases.head()
```

```python
df_closed_cases['count'].fillna(df_closed_cases['count'].median(),inplace=True)
```

```python
# Number of closed cases per state
fig, ax = plt.subplots(1, figsize=(10,6))
df_closed_cases.plot(column='count',cmap='Blues', linewidth=1, ax=ax, edgecolor='0.9', legend = True)
ax.axis('off')
ax.set_title("Number of closed cases per state - 2011")
```

**CASES UNCLOSED PER STATE**

```python
df_unclosed = pd.DataFrame({'state_name':df_status_count['state_name'],'count':df_status_count.loc[df_status_count['status'] == 'not closed']['count']})
df_unclosed = df_unclosed.dropna(subset=['count'], how='all')
df_unclosed.head()
```

```python
df_unclosed_cases = pd.merge(map_df,df_unclosed,on='state_name', how='left')
df_unclosed_cases=df_unclosed_cases.drop(columns=['ID_0', 'ISO', 'NAME_0', 'ID_1', 'HASC_1', 'CCN_1',
        'CCA_1', 'TYPE_1', 'ENGTYPE_1', 'NL_NAME_1', 'VARNAME_1'])
df_unclosed_cases.head()
```

```python
df_closed_cases['count'].fillna(df_closed_cases['count'].median(),inplace=True)
```

```python
# Number of unclosed cases per state
fig, ax = plt.subplots(1, figsize=(10,6))
df_unclosed_cases.plot(column='count',cmap='Blues', linewidth=1, ax=ax, edgecolor='0.9', legend = True)
ax.axis('off')
ax.set_title("Number of unclosed cases per state - 2011")
```

**NUMBER OF COURTS PER STATE**

```python
df_courts_num = pd.DataFrame({'state_name':df['state_name'],'court_no':df['court_no']})
df_courts_num = df_courts_num.drop_duplicates()

df_courts_num_final = pd.DataFrame({'state_name':df_courts_num.groupby(['state_name'])['court_no'].sum().index.values,'count':df_courts_num.groupby(['state_name'])['court_no'].sum().values})
df_courts_num_final.head()
```

```python
df_courts = pd.merge(map_df,df_courts_num_final,on='state_name', how='left')
df_courts=df_courts.drop(columns=['ID_0', 'ISO', 'NAME_0', 'ID_1', 'HASC_1', 'CCN_1',
        'CCA_1', 'TYPE_1', 'ENGTYPE_1', 'NL_NAME_1', 'VARNAME_1'])
df_courts.head()
```

```python
df_courts['count'].fillna(df_courts['count'].median(),inplace=True)
```

```python
# Number of unclosed cases per state
fig, ax = plt.subplots(1, figsize=(10,6))
df_courts.plot(column='count',cmap='Blues', linewidth=1, ax=ax, edgecolor='0.9', legend = True)
ax.axis('off')
ax.set_title("Number of courts per state - 2011")
```

**CLOSED AND UNCLOSED CASES PER PENDING CASES PER STATE**

```python
df_unclosed = df_unclosed.rename(columns={'count':'unclosed_count'})
df_closed = df_closed.rename(columns={'count':'closed_count'})
df_courts_cases = pd.merge(df_courts_num_final, df_unclosed,on = 'state_name', how = 'left')
df_courts_cases = pd.merge(df_courts_cases, df_closed,on = 'state_name', how = 'left')
df_courts_cases.head()
```

```python
df_courts_cases['unclosed_avg_count'] = df_courts_cases['unclosed_count'] / df_courts_cases['count']
df_courts_cases['closed_avg_count'] = df_courts_cases['closed_count'] / df_courts_cases['count']
df_courts_cases.head()
```

```python
fig_unclosed_avg = px.bar(df_courts_cases, x='state_name', y='unclosed_avg_count', title='Average unclosed cases per court per state - 2011')
fig_unclosed_avg.show()
```

```python
fig_unclosed_avg = px.bar(df_courts_cases, x='state_name', y='closed_avg_count', title='Average closed cases per court per state - 2011')
fig_unclosed_avg.show()
```
